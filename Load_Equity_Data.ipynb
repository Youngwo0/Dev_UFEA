```python
import yfinance as yf
import pandas as pd
```

```python
df_aapl = yf.download("AAPL", start="2021-01-01", end="2021-12-31")
df_aapl
```

```python
df_msft = yf.download("MSFT", start="2021-01-01", end="2021-12-31")
df_msft
```

```python
df_aapl_msft = pd.concat([df_aapl, df_msft], axis=1)
```

```python
df_aapl_msft.to_csv('aapl_msft.csv')
```
